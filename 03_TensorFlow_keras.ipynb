```python
from tensorflow.keras.models import Sequential    # Sequencial Interface
from tensorflow.keras.layers import Dense   # Fully Connected layer
```

```python
model = Sequential([
    
    Dense(4,
          activation= "relu",
          input_shape=(2,),
          kernel_initializer='glorot_uniform',name= "Hidden_layer"),
 
    Dense(3,
          activation= "softmax",
          kernel_initializer='glorot_uniform',name="Output_layer" )
    
])
```

###  During backpropagation, which of the following activation functions is/are not suitable for hidden layers if we wish to avoid the problem of vanishing gradient?

Correct option :

- Tanh
- Sigmoid
- Binary step

##### Explanation :


- Backpropagation calculates the gradient of the loss function to update the parameters of the neural network.
- Tanh and sigmoid functions suffer from vanishing gradient problems when the gradient approaches zero as more and more layers are used in backpropagation. 
https://machinelearningmastery.com/how-to-fix-vanishing-gradients-using-the-rectified-linear-activation-function/
- The binary step function is not used because its derivative at all points is zero, which does not help in backpropagation.
